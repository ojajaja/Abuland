# Exploratory Data Analysis (EDA)

### Data content

1. **Year**: The year of the data record.
2. **RiceCons**: Domestic consumption of rice in Singapore in thousand metric tons.
3. **PerCapGNI**: Gross National Income per capita.
4. **PerCapGDP**: Gross Domestic Product per capita.
5. **GNI_USD**: Gross National Income per capita in US dollars.
6. **GDP_USD**: Gross Domestic Product per capita in US dollars.
7. **Inflation**: Annual inflation rate based on consumer prices.
8. **MeanPriceUSD**: Mean price of rice per metric ton in USD.
9. **TotPop**: Total population of Singapore.
10. **ResPop**: Number of residents in Singapore.
11. **CitizenPop**: Number of Singaporean citizens.
12. **PRPop**: Number of permanent residents in Singapore.
13. **NonResPop**: Number of non-residents in Singapore.
14. **PopGrowth**: Growth rate of the total population in percentage.
15. **ResPopGrowth**: Growth rate of the resident population in percentage.
16. **PopDensity**: Population density in Singapore, measured as the total population per square kilometer.
17. **SexRatio**: Number of males per thousand females.
18. **MedAgeRes**: Median age of the resident population.
19. **MedAgeCit**: Median age of the citizen population.
20. **OldSupRatio**: Ratio of residents aged 15-64 years to those aged 65 years and over.
21. **AgeDepRatio**: Ratio of residents aged under 15 years and 65 years and over per hundred residents aged 15-64 years.
22. **ChildDepRat**: Ratio of residents aged under 15 years per hundred residents aged 15-64 years.
23. **OldDepRatio**: Ratio of residents aged 65 years and over per hundred residents aged 15-64 years.
24. **OldSup20_64**: Ratio of residents aged 20-64 years to those aged 65 years and over.
25. **AgeDep20_64**: Ratio of residents aged under 20 years and 65 years and over per hundred residents aged 20-64 years.
26. **ChildDep20_64**: Ratio of residents aged under 20 years per hundred residents aged 20-64 years.
27. **OldDep20_64**: Ratio of residents aged 65 years and over per hundred residents aged 20-64 years.
28. **CitSup15_64**: Ratio of citizens aged 15-64 years to those aged 65 years and over.
29. **CitDep15_64**: Ratio of citizens aged under 15 years and 65 years and over per hundred citizens aged 15-64 years.
30. **CitChildDep15_64**: Ratio of citizens aged under 15 years per hundred citizens aged 15-64 years.
31. **CitOldDep15_64**: Ratio of citizens aged 65 years and over per hundred citizens aged 15-64 years.
32. **CitSup20_64**: Ratio of citizens aged 20-64 years to those aged 65 years and over.
33. **CitDep20_64**: Ratio of citizens aged under 20 years and 65 years and over per hundred citizens aged 20-64 years.
34. **CitChildDep20_64**: Ratio of citizens aged under 20 years per hundred citizens aged 20-64 years.
35. **CitOldDep20_64**: Ratio of citizens aged 65 years and over per hundred citizens aged 20-64 years.
36. **ResNatInc**: Natural increase of the resident population (births minus deaths).
37. **NatIncRate**: Rate of natural increase per thousand residents.
38. **IndPrecip**: Precipitation in India in millimeters.
39. **IndTemp**: Temperature in India in degrees Celsius.
40. **ThaPrecip**: Precipitation in Thailand in millimeters.

#### Skewness and Kurtosis

```python
# Lists to store skewness and kurtosis
skewness = []
kurtosis = []
columns = []

# Calculate skewness and kurtosis for each column
for col in df.columns:
    sk = df[col].astype(float).skew()
    kt = df[col].astype(float).kurt()
    
    # Append the values and column name
    skewness.append(sk)
    kurtosis.append(kt)
    columns.append(col)

# Create a DataFrame to hold the skewness and kurtosis values
sk_kt_df = pd.DataFrame({'Skewness': skewness, 'Kurtosis': kurtosis}, index=columns)

# Plotting
plt.figure(figsize=(14, 6))

# Skewness plot
plt.subplot(1, 2, 1)
sns.barplot(x=sk_kt_df.index, y=sk_kt_df['Skewness'], palette='viridis')
plt.title('Skewness of Variables')
plt.xticks(rotation=45, ha='right')  # Rotate labels for better readability

# Kurtosis plot
plt.subplot(1, 2, 2)
sns.barplot(x=sk_kt_df.index, y=sk_kt_df['Kurtosis'], palette='viridis')
plt.title('Kurtosis of Variables')
plt.xticks(rotation=45, ha='right')  # Rotate labels for better readability

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['RiceCons'], kde=True)
plt.title('Distribution of Rice Domestic Consumption')
plt.show()
```

```python
# Drop 'Year' column from the dataframe
df_without_year = df.drop(columns=['Year'])

# Compute the correlation matrix
correlation_matrix = df_without_year.corr()

# Plot the heatmap
plt.figure(figsize=(20, 10))
sns.heatmap(correlation_matrix, annot = True, fmt = ".1f", linewidths = .7)
plt.title('Correlation Matrix')
plt.show()
```

### Kernel Density Plot

```python
# Extract the correlation with the target variable
target_correlation = correlation_matrix['RiceCons'].sort_values(ascending=False)

# Display the correlation
print(target_correlation)
# Select top 10 features most correlated with RiceCons
top_features = target_correlation.index[1:11].tolist()
top_features.append('RiceCons')

# Create a new DataFrame with the selected features
df_top_features = df[top_features]
```

```python
df_top_features.head()
```

```python
# Function to plot KDE for each feature against the target variable
def plot_kde(feature, ax):
    sns.kdeplot(data=df, x=feature, y='RiceCons', fill=True, cmap="viridis", ax=ax)
    ax.set_title(f'KDE Plot of {feature} vs RiceCons')
    ax.set_xlabel(feature)
    ax.set_ylabel('RiceCons')

# Number of features to plot
num_features = len(top_features) - 1  # Exclude 'RiceCons'
num_cols = 3  # Number of plots per row
num_rows = (num_features + num_cols - 1) // num_cols  # Calculate the number of rows needed

fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 5 * num_rows))
axes = axes.flatten()  # Flatten the axes array for easy iteration

# Plot KDE for each of the top correlated features
for i, feature in enumerate(top_features[:-1]):  # Exclude 'RiceCons' from the loop
    plot_kde(feature, axes[i])

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

### Simple Linear Regression

```python
# Function to perform simple linear regression and plot the results
def simple_linear_regression(feature, ax):
    X = df[[feature]].dropna()
    y = df['RiceCons'].loc[X.index]
    
    model = LinearRegression()
    model.fit(X, y)
    
    y_pred = model.predict(X)
    r2 = r2_score(y, y_pred)
    
    sns.scatterplot(x=X[feature], y=y, ax=ax)
    ax.plot(X[feature], y_pred, color='red')
    ax.set_title(f'Simple Linear Regression: {feature} vs RiceCons\nRÂ² = {r2:.2f}')
    ax.set_xlabel(feature)
    ax.set_ylabel('RiceCons')

# Number of features to plot
num_features = len(top_features) - 1  # Exclude 'RiceCons'
num_cols = 3  # Number of plots per row
num_rows = (num_features + num_cols - 1) // num_cols  # Calculate the number of rows needed

fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 5 * num_rows))
axes = axes.flatten()  # Flatten the axes array for easy iteration

# Plot simple linear regression for each of the top correlated features
for i, feature in enumerate(top_features[:-1]):  # Exclude 'RiceCons' from the loop
    simple_linear_regression(feature, axes[i])

# Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
